```python
# Import pandas library for data analysis
import pandas as pd

# Load monthly Divvy trip data (2022)

# January to August

df_202201 = pd.read_csv("/content/202201-divvy-tripdata.csv")
df_202202 = pd.read_csv("/content/202202-divvy-tripdata.csv")
df_202203 = pd.read_csv("/content/202203-divvy-tripdata.csv")
df_202204 = pd.read_csv("/content/202204-divvy-tripdata.csv")
df_202205 = pd.read_csv("/content/202205-divvy-tripdata.csv")
df_202206 = pd.read_csv("/content/202206-divvy-tripdata.csv")
df_202207 = pd.read_csv("/content/202207-divvy-tripdata.csv")
df_202208 = pd.read_csv("/content/202208-divvy-tripdata.csv")

# September (corrupted → skip bad lines) finnally data haven't bad line
df_202209 = pd.read_csv("/content/202209-divvy-tripdata.csv")

df_202210 = pd.read_csv("/content/202210-divvy-tripdata.csv")
df_202211 = pd.read_csv("/content/202211-divvy-tripdata.csv")
df_202212 = pd.read_csv("/content/202212-divvy-tripdata.csv")
```

```python
# Combine all monthly dataframes into one dataset

df_all = pd.concat([
    df_202201, df_202202, df_202203, df_202204,
    df_202205, df_202206, df_202207, df_202208,
    df_202209, df_202210, df_202211, df_202212
], ignore_index=True)

# Display summary information about the combined dataset

df_all.info()
```

```python
# Convert 'started_at' and 'ended_at' columns to datetime format
# 'format="mixed"' allows different date formats in the same column
# 'errors="coerce"' converts invalid or corrupted values to NaT (missing datetime)

df_all['started_at'] = pd.to_datetime(df_all['started_at'], format='mixed', errors='coerce')
df_all['ended_at'] = pd.to_datetime(df_all['ended_at'], format='mixed', errors='coerce')

# Display the first few values of 'started_at' to verify conversion

df_all['started_at'].head()
```

```python
# Create a new column 'ride_length' by calculating the difference
# between 'ended_at' and 'started_at'

df_all['ride_length'] = df_all['ended_at'] - df_all['started_at']

# Display the first few values to verify the result

df_all['ride_length'].head()
```

```python
# Sort the dataframe by 'ride_length' in descending order
# This helps identify the longest trips in the dataset

df_all.sort_values('ride_length', ascending=False).head(10)
```

```python
# Remove rides that are zero or negative
df_all = df_all[df_all['ride_length'] > pd.Timedelta(0)]
```

```python
# Remove rides longer than 24 hours
df_all = df_all[df_all['ride_length'] < pd.Timedelta(days=1)]
```

```python
#Verify if Remove rides longer than 24 hours
df_all.sort_values('ride_length', ascending=False).head(10)
```

```python
# Create a new column 'day_of_week' from the 'started_at' datetime
# This extracts the full weekday name (e.g., Monday, Tuesday)

df_all.loc[:, 'day_of_week'] = df_all['started_at'].dt.day_name()

# Display the first few values to verify the new column

df_all.loc[:, 'day_of_week'].head()
```

```python
#Average ride length by user type (members vs casual)

# Group data by 'member_casual' and calculate the mean ride length

df_all.groupby('member_casual')['ride_length'].mean()
```

```python
# Calculate the average ride length for each day of the week

df_all.groupby('day_of_week')['ride_length'].mean()
```

```python
#Count number of member and cusual
df_all.groupby('member_casual')['ride_length'].count()
```

```python
#Number of rides per day of the week

# Group data by user type (member vs casual) and day of the week
# Count the number of rides using 'ride_id'

df_all.groupby(['member_casual', 'day_of_week'])['ride_id'].count()
```

```python
#Or cleaner:
df_all['day_of_week'] = df_all['started_at'].dt.day_name()

df_all.groupby(['member_casual', 'day_of_week']).size()
```

```python
#Number of rides per month

# Create a 'month' column from the 'started_at' datetime
# This extracts the full month name (e.g., January, February)

df_all['month'] = df_all['started_at'].dt.month_name()

# Group data by user type (member vs casual) and month
# Count the number of rides in each group

df_all.groupby(['member_casual', 'month']).size()

```

```python
#Bike type usage by user type

# Group data by user type ('member_casual') and bike type ('rideable_type')
# Count the number of rides in each group

df_all.groupby(['member_casual', 'rideable_type']).size()
```

```python
# Peak hours analysis

# Create a new column 'hour' from 'started_at' datetime
# This extracts the hour of the day (0–23) when the ride started

df_all['hour'] = df_all['started_at'].dt.hour

# Group data by user type ('member_casual') and hour
# Count the number of rides in each group to find peak usage times

df_all.groupby(['member_casual', 'hour']).size()
```

```python
# Import visualization libraries

import seaborn as sns
import matplotlib.pyplot as plt

# Plot number of rides per day of the week, separated by user type

plt.figure(figsize=(12,6))   # Set figure size

sns.countplot(data=df_all, x='day_of_week', hue='member_casual',
              order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']    # Ensure correct day order
)

# Add titles and axis labels
                   )
plt.title("Rides per Day of Week: Members vs Casual")
plt.xlabel("Day of Week")
plt.ylabel("Number of Rides")

# Show the plot

plt.show()
```

```python
# Convert ride length to minutes for easier interpretation

df_all['ride_length_minutes'] = df_all['ride_length'].dt.total_seconds() / 60

# Plot average ride length by user type

plt.figure(figsize=(8,6))   # Set figure size

sns.barplot(data=df_all, x='member_casual', y='ride_length_minutes')

# Add titles and axis labels

plt.title("Average Ride Length (Minutes)")
plt.xlabel("Rider Type")
plt.ylabel("Average Ride Length (min)")

# Show the plot
plt.show()
```

```python
# Extract the month name from the 'started_at' datetime
# This creates a new column 'month' for month-level analysis

df_all['month'] = df_all['started_at'].dt.month_name()
```

```python
# Plot number of rides per month, separated by user type (member vs casual)

plt.figure(figsize=(12,6))   # Set figure size

sns.countplot(data=df_all, x='month', hue='member_casual',
              order=['January','February','March','April','May','June','July','August','September','October','November','December']    # Ensure chronological order     )

# Add titles and axis labels

plt.title("Rides per Month: Seasonality")
plt.xlabel("Month")
plt.ylabel("Number of Rides")
plt.xticks(rotation=45)     # Rotate x-axis labels for readability

# Show the plot
plt.show()
```

```python
# Peak hours analysis: visualize rides by hour of day and user type
# Extract the hour from 'started_at' datetime for each ride

df_all['hour'] = df_all['started_at'].dt.hour

# Plot number of rides by hour, separated by user type

plt.figure(figsize=(12,6))    # Set figure size

sns.countplot(data=df_all, x='hour', hue='member_casual')

# Add titles and axis labels

plt.title("Peak Hours of Usage")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Rides")

# Show the plot
plt.show()
```

```python
# Bike type usage: visualize rides by bike type and user type

plt.figure(figsize=(8,6))   # Set figure size

sns.countplot(data=df_all, x='rideable_type', hue='member_casual')

# Add titles and axis labels

plt.title("Bike Type Usage")
plt.xlabel("Bike Type")
plt.ylabel("Number of Rides")

# Show the plot

plt.show()
```
